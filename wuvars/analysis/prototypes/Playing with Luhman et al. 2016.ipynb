# Playing with Luhman et al. 2016

The paper [ A Census of Young Stars and Brown Dwarfs in IC 348 and NGC 1333 ](https://ui.adsabs.harvard.edu/abs/2016ApJ...827...52L/abstract) seems to be, potentially, an amazing one-stop-shop option for brown dwarfs in these two regions. Let's assess how likely this might be - is this a replacement for having to cobble together everything from (e.g.) all the Scholz papers, Rebull 2015, etc etc. Now, let's look at the contents (the tables!) and see if this is where I'll get my spectral types & all that.

```python
import astropy.table
import matplotlib.pyplot as plt
```

```python
from wuvars.analysis.spectral_type_to_number import get_num_from_SpT
```

```python
# it lives...
aux_path = "/Users/tsrice/Documents/Variability_Project_2020/wuvars/data/auxiliary_catalogs/IC348"
filepath = os.path.join(aux_path, "Luhman2016_Table1_apjaa2cabt1_mrt.txt")

# load up our data.
table = astropy.table.Table.read(filepath, format="ascii.cds")
```

```python
table
```

```python
spt_list = []

for item in table['Adopt']:
    if np.ma.is_masked(item):
        print(item)
        spt_list.append(np.nan)
    else:
        print(item,'\t', get_num_from_SpT(item))
        spt_list.append(get_num_from_SpT(item))
```

```python
plt.plot(np.array(spt_list), 'k.')
plt.axhline(6)
```

```python
is_brown_dwarf = (np.array(spt_list) >= 6)
print(np.sum(is_brown_dwarf))
```

```python
np.where(is_brown_dwarf & (table['Ksmag'] < 12))
```

```python
table[189]
```

```python
t = table
ra_dd = (t['RAh'] + t['RAm']/60 + t['RAs']/3600) * 360/24
dec_dd = (t['DEd'] + t['DEm']/60 + t['DEs']/3600) 
    
```

```python
plt.figure(figsize=(5,5))
plt.plot(ra_dd, dec_dd, 'k.')
plt.plot(ra_dd[is_brown_dwarf], dec_dd[is_brown_dwarf], 'mo', label="M6 or later")
plt.gca().invert_xaxis()
plt.title("IC 348")
plt.legend()
```
