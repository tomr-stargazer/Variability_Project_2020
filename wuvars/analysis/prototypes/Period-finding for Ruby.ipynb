# The goal

Ruby has been working on analyzing the variability of a number of objects in Perseus.

We will today analyze the variables' light curves for periodic signals.

```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.timeseries import LombScargle
from wuvars.analysis.periods import f_max, f_min, N_eval
from wuvars.plotting.lightcurve import simple_phased_lc_scatter_gridspec
from wuvars.plotting.lightcurve import simple_lc_scatter
from wuvars.plotting.lightcurve import plot_phase_core
```

```python
# 1. Load the data and plot it as lightcurves.

# times, mags, errs = np.loadtxt("some_file_from_ruby.txt")

times = np.loadtxt("ruby_star_periods/44508746117813MJD")
mags = np.loadtxt("ruby_star_periods/44508746117813KMAG")
errs = np.loadtxt("ruby_star_periods/44508746117813KMAGERR")

plt.errorbar(times, mags, yerr=errs, fmt='k.')
plt.gca().invert_yaxis()

```

```python
# 2. Compute and plot periodograms.

# for item in list: # pseudocode
    
    # do something like this...

freq = np.linspace(f_min, f_max, N_eval)

ls = LombScargle(times, mags, dy=errs)
power = ls.power(freq, assume_regular_frequency=True)


min_freq = 1 / 100
power[freq < min_freq] = 0
# power[np.abs(freq-1) < 0.01] = 0

powermax = np.nanmax(power)
fmax = freq[np.nanargmax(power)]
fap = ls.false_alarm_probability(np.nanmax(power))

amp = np.sqrt(np.sum(ls.model_parameters(fmax)[1:3]**2))

print(fap)
print(fmax)

fig, ax = plt.subplots()
plot_phase_core(ax, times, mags, errs, 1/fmax)
plt.gca().invert_yaxis()

```

```python
plt.figure()
plt.plot(1 / freq, power, "k", rasterized=True)
plt.semilogx()
plt.xlim(0.1, 10)
```
