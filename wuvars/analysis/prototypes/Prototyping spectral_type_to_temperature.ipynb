```python
L_spt_nums = []
L_teffs = []

for (L_spt, L_teff) in Faherty_spt_teff_pairs:
    L_spt_nums.append(int(L_spt[1])+10)
    L_teffs.append(L_teff)
```

```python
print(L_spt_nums)
```

```python
all_spt_nums = []
all_teffs = []

for (spt, teff) in all_spt_teff_pairs:
    if spt[0] == 'M':
        all_spt_nums.append(int(spt[1]))
    else:
        all_spt_nums.append(int(spt[1])+10)            
    all_teffs.append(teff)
```

```python
fig, ax = plt.subplots(1, figsize=(6,5))
plt.plot(spt_nums, teffs, 'k.-', label='M spectral type-temperature relation\n(Luhman+03)')
plt.plot(L_spt_nums, L_teffs, 'm.-', label='L spectral type-temperature relation\n(Faherty+16)')
plt.plot(all_spt_nums, all_teffs, 'k:', lw=0.5)

plt.axvline(6, lw=0.5, ls="--", alpha=0.3)
plt.axhline(2990, lw=0.5, ls="--", alpha=0.3)

plt.xlabel("Spectral type (M, L)")
plt.ylabel("Teff (K)")
plt.legend()

plt.title(r"Spectral type-$T_{\rm{eff}}$ relation for M and L dwarfs")
```
