# SIDSEP: a standard string format for sourceids

WFCAM/UKIRT assigns all sources a 14-digit SOURCEID. It is really, really annoying to try adn read or write 14 digits.

Bo had the excellent idea (implicitly communicated to me) to separate the number (like a phone number!) into clusters, turning (e.g.)
`44027709751316`
into
`4402 7709 7513 16`.


```python
from wuvars.data.spreadsheet import v1 as spread
```

```python
spread[0]
```

```python
%run ../sidsep.py
```

```python
import textwrap
```

```python
sid = 44027709751316
sidsep(sid)
```

```python
for i in range(100):
    sid = spread[0].index[i]
    sep = sidsep(sid)
    print(f"{sid}: {sep}")
```
