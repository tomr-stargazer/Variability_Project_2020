```python
coords = SkyCoord(
    ra=bd_results_table["RA"] * u.rad, dec=bd_results_table ["DEC"] * u.rad
)
```

```python
latex_strings = []

for i, row in enumerate(bd_results_table):
#     latex_items = []
    
    name = row['Name']
    ra = coords[i].ra.to_string(sep=" ", unit=u.hourangle, precision=2, pad=True)
    dec = coords[i].dec.to_string(sep=" ", precision=1, alwayssign=True)
    
    j = row['JAPERMAG3']
    h = row['HAPERMAG3']
    k = row['KAPERMAG3']
    j_e = row['JAPERMAG3ERR']
    h_e = row['HAPERMAG3ERR']
    k_e = row['KAPERMAG3ERR']    
    
    j_pm = f"{j:.2f} $\pm$ {j_e:.2f}"
    h_pm = f"{h:.2f} $\pm$ {h_e:.2f}"
    k_pm = f"{k:.2f} $\pm$ {k_e:.2f}"
    
    SpT = row['SpT']
    
    ref = '1'
    
    latex_items = [name, ra, dec, j_pm, h_pm, k_pm, SpT, ref]
    
    latex_row_string = ' & '.join(latex_items)+r' \\'
    
    print(latex_row_string)
    
    
```
